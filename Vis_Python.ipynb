```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets 
```

```python
data = sns.load_dataset('penguins')
data.head()
```

```python
# Histogram
sns.histplot(data = data, x = 'body_mass_g')
plt.title("Distribution of Body Mass")
plt.xlabel("Body Mass")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot
sns.scatterplot(data = data, x='bill_length_mm', y='flipper_length_mm', hue='species')
plt.title('Scatter Plot : Bill Length vs. Flipper Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.show()
```

```python
# Bar Plot
sns.barplot(data = data, x = 'species', y = 'body_mass_g', hue = 'species')
plt.title('Bar Plot : Bill Length vs. Body Mass')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Body Mass')
plt.show()
```

```python
# BoxPlot
sns.boxplot(data=data, x='species', y='flipper_length_mm')
plt.title('Box : Flipper Length by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.tight_layout()
plt.show()
```

```python
# Pair Plot
sns.pairplot(data = data, hue = 'species')
```

```python
# Correlation Matrix
temp = data.iloc[:, 2:5]
matrix = temp.corr()
sns.heatmap(matrix, annot = True)
```

```python
# Violin Plot
sns.violinplot(data=data, x='species', y='flipper_length_mm')
plt.title('Box : Flipper Length by Species')
plt.xlabel('Species')
plt.ylabel('Flipper Length (mm)')
plt.tight_layout()
plt.show()
```

```python
# CountPlot
sns.countplot(data=data, x='island', hue='sex')
plt.title('Count of Male and Female Penguins per Island')
plt.xlabel('Island')
plt.ylabel('Count')
plt.legend(title='Sex')
plt.tight_layout()
plt.show()
```

```python
# Line Plot
sns.lineplot(data = data, x=data['flipper_length_mm'],y=data['body_mass_g'])
```
